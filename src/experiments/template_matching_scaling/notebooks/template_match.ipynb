```python
%matplotlib inline

# We have to add a reference to the parent folder in order to access the modules.
import sys
sys.path.insert(1, os.path.join(sys.path[0], '..'))

import template_matching_scaling as templ
from tseenum import enum
import matplotlib as mpl
import matplotlib.pyplot as plt
import cv2
import math

mpl.rcParams['figure.figsize'] = (16.5, 15.0)
    
image_path = "../../eval_data/motion_images/wiltshire_outside_10cm"

config_file = "../../perspective_calibration/data/calibdata_23_03_15_11_07_04.txt" 

match_types = enum('DISTANCE', 'DISTANCE_ED', 'HIST')

image_pairs = [("IMG1.JPG", "IMG2.JPG"),("IMG3.JPG", "IMG4.JPG"),("IMG5.JPG", "IMG6.JPG")]

plot_count = len(image_pairs)

column_max = 2

row_max = int(math.ceil(plot_count / float(column_max)))

fig, axes = plt.subplots(row_max, column_max)

column_count = 0
row_count = 0

for pair in image_pairs:
    image_1_path = "{0}/{1}".format(image_path, pair[0])
    image_2_path = "{0}/{1}".format(image_path, pair[1])
    
    match_method1 = templ.MatchMethod("DistanceEuclidean", match_types.DISTANCE_ED, None, "r")
    match_method2 = templ.MatchMethod("HistCorrel", match_types.HIST, cv2.cv.CV_COMP_CORREL, "b", True)
    match_method3 = templ.MatchMethod("HistChiSqr", match_types.HIST, cv2.cv.CV_COMP_CHISQR, "g")
    
    if row_max > 1:
        match = templ.TemplateMatching(image_1_path, image_2_path, config_file, match_types, axes[row_count, column_count])

    else:
        match = templ.TemplateMatching(image_1_path, image_2_path, config_file, match_types, axes[column_count])
        
    match.search_image(100, [match_method1, match_method2, match_method3])
    
    if column_count == (column_max - 1):
        row_count += 1
        column_count = 0
    else:
        column_count += 1
        
# If we do not have an even number of graphs, then we need to remove the last blank one.
if (plot_count % column_max) != 0: 
    
    if row_max > 1:
        
        axes[-1, -1].axis('off')
        
    else:
        
        axes[-1].axis('off')
    
plt.show()
```
