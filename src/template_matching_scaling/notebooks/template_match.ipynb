# Test 1: Wiltshire Outside

```python
%matplotlib inline

# We have to add a reference to the parent folder in order to access the modules.
import sys,os
sys.path.insert(1, os.path.join(sys.path[0], '..'))

import template_matching_scaling as templ
import matplotlib as mpl
import matplotlib.pyplot as plt
import cv2

from tse.tse_match_methods import tse_match_methods

mpl.rcParams['figure.figsize'] = (16.5, 15.0)

image_path = "../../eval_data/motion_images/wiltshire_outside_10cm"

config_file = "../../perspective_calibration/data/calibdata_23_03_15_11_07_04.txt"

image_pairs = [("IMG1.JPG", "IMG2.JPG"), ("IMG3.JPG", "IMG4.JPG"), ("IMG5.JPG", "IMG6.JPG"), ("IMG7.JPG", "IMG8.JPG")]

patch_sizes = [100, 200]

match_method1 = templ.MatchMethod("DistanceEuclidean", tse_match_methods.DISTANCE_ED, None, "r")
match_method2 = templ.MatchMethod("HistCorrel", tse_match_methods.HIST, cv2.cv.CV_COMP_CORREL, "b", True)
match_method3 = templ.MatchMethod("HistChiSqr", tse_match_methods.HIST, cv2.cv.CV_COMP_CHISQR, "g")

match_methods = [match_method1, match_method2, match_method3]

templ.start_tests(image_path, image_pairs, patch_sizes, match_methods, config_file)
```
