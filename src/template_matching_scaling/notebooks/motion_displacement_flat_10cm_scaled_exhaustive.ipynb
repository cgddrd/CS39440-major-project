## 2) Calculate Displacement 

```python
%%cache cache_flat_10cm_scaled_exhaustive.pkl results_dict

config_file = "../../perspective_calibration/data/calibdata_23_03_15_11_07_04.txt"

image_pairs = [
                ("../../eval_data/motion_images/flat_10cm/IMG1.JPG", "../../eval_data/motion_images/flat_10cm/IMG2.JPG"),
                ("../../eval_data/motion_images/flat_10cm/IMG3.JPG", "../../eval_data/motion_images/flat_10cm/IMG4.JPG"),
                ("../../eval_data/motion_images/flat_10cm/IMG5.JPG", "../../eval_data/motion_images/flat_10cm/IMG6.JPG"),
                ("../../eval_data/motion_images/flat_10cm/IMG7.JPG", "../../eval_data/motion_images/flat_10cm/IMG8.JPG"),
                ("../../eval_data/motion_images/flat_10cm/IMG9.JPG", "../../eval_data/motion_images/flat_10cm/IMG10.JPG"),
                ("../../eval_data/motion_images/flat_10cm/IMG11.JPG", "../../eval_data/motion_images/flat_10cm/IMG12.JPG")
              ]

patch_sizes = [100, 200]

match_method1 = TSEMatchType("DistanceEuclidean", tse_match_methods.DISTANCE_ED, None, "r", reverse_score=True)
match_method2 = TSEMatchType("HistCorrel", tse_match_methods.HIST, cv2.cv.CV_COMP_CORREL, "b")
match_method3 = TSEMatchType("HistChiSqr", tse_match_methods.HIST, cv2.cv.CV_COMP_CHISQR, "g", reverse_score=True)
match_method4 = TSEMatchType("DistanceCorr", tse_match_methods.DISTANCE, cv2.cv.CV_TM_CCORR_NORMED, "y")

match_methods = [match_method1, match_method2, match_method3, match_method4]

results_dict = templ.start_tests(image_pairs, patch_sizes, match_methods, config_file, exhaustive_search=True, use_scaling=True, plot_results=False)
```

### Test 4: IMG1 - IMG12 - 100px patch - *DistanceEuclidean, DistanceCorr, HistCorrel, HistChiSqr*

```python
results = results_dict

final_results = {}

format_ids = ["b", "g", "r", "y"]

for image_pair in results:
    
    for method in results[image_pair][100]:
        
        if method not in final_results:
            final_results[method] = {}
        
        data = results[image_pair][100][method]
        
        if 'displacement' in final_results[method]:
            final_results[method]['displacement'].append([o.displacement for o in data])
        else:
            final_results[method]['displacement'] = [[o.displacement for o in data]]
            
        final_results[method]['row'] = [o.row for o in data]
        

for i, result in enumerate(final_results):
    
    final_results[result]['average'] = TSEDataUtils.calc_element_wise_average(final_results[result]['displacement'])
    
    final_results[result]['filtered'] = TSEDataUtils.filter_outliers_ab_dist_median(final_results[result]['average'])
    
    final_results[result]['filtered_row'] = np.array(final_results[result]['row'])[TSEDataUtils.filter_outliers_ab_dist_median_indices(final_results[result]['average'])]
    
    final_results[result]['cma'] = TSEDataUtils.calc_centered_moving_average(np.array(final_results[result]['filtered']), 10)
    
    plt.plot(np.array(final_results[result]['filtered_row']), np.array(final_results[result]['filtered']), "{0}.".format(format_ids[i]), label="{0} (Filtered)".format(result))
    
    plt.plot(final_results[result]['filtered_row'][len(final_results[result]['filtered_row']) - len(final_results[result]['cma']):], final_results[result]['cma'], "{0}-".format(format_ids[i]), label="{0}_CMA".format(result))                                                                                         
          
plt.title('Average Vertical Motion Displacement - IMG1-IMG12 - 100px patch height')
plt.xlabel('Image Row Height (px)')
plt.ylabel('Average Vertical Displacement (px)')
plt.legend(loc='upper left')
plt.show()
```

```python
results = results_dict

raw_results_histcorrel = []
raw_results_ed = [] 
raw_results_histchisqr = [] 
raw_results_corr = []

image_rows_histcorrel = []
image_rows_ed = []
image_rows_histchisqr = []
image_rows_corr = []

final_results = {}

format_ids = ["b", "g", "r", "y"]

plt.title('Average Vertical Motion Displacement - IMG1-IMG12 - 100px patch height')
plt.xlabel('Image Row Height (px)')
plt.ylabel('Average Vertical Displacement (px)')
plt.legend(loc='upper left')

for image_pair in results:    
    data_histcorrel = results[image_pair][100]['HistCorrel']
    data_histchisqr = results[image_pair][100]['HistChiSqr']
    data_ed = results[image_pair][100]['DistanceEuclidean']
    data_corr = results[image_pair][100]['DistanceCorr']
    
    raw_results_histcorrel.append([o.displacement for o in data_histcorrel])
    raw_results_ed.append([o.displacement for o in data_ed])
    raw_results_histchisqr.append([o.displacement for o in data_histchisqr])
    raw_results_corr.append([o.displacement for o in data_corr])
    
    image_rows_histcorrel = [o.row for o in data_histcorrel]
    image_rows_ed = [o.row for o in data_ed]
    image_rows_histchisqr = [o.row for o in data_histchisqr]
    image_rows_corr = [o.row for o in data_corr]
    
averaged_results_histcorrel = TSEDataUtils.calc_element_wise_average(raw_results_histcorrel)
averaged_results_ed = TSEDataUtils.calc_element_wise_average(raw_results_ed)
averaged_results_histchisqr = TSEDataUtils.calc_element_wise_average(raw_results_histchisqr)
averaged_results_corr = TSEDataUtils.calc_element_wise_average(raw_results_corr)

filtered_results_histcorrel = TSEDataUtils.filter_outliers_ab_dist_median(averaged_results_histcorrel)
filtered_results_ed = TSEDataUtils.filter_outliers_ab_dist_median(averaged_results_ed)
filtered_results_histchisqr = TSEDataUtils.filter_outliers_ab_dist_median(averaged_results_histchisqr)
filtered_results_corr = TSEDataUtils.filter_outliers_ab_dist_median(averaged_results_corr)

image_rows_histcorrel = np.array(image_rows_histcorrel)[TSEDataUtils.filter_outliers_ab_dist_median_indices(averaged_results_histcorrel)]
image_rows_ed = np.array(image_rows_ed)[TSEDataUtils.filter_outliers_ab_dist_median_indices(averaged_results_ed)]
image_rows_histchisqr = np.array(image_rows_histchisqr)[TSEDataUtils.filter_outliers_ab_dist_median_indices(averaged_results_histchisqr)]
image_rows_corr = np.array(image_rows_corr)[TSEDataUtils.filter_outliers_ab_dist_median_indices(averaged_results_corr)]

y_centered_moving_average_histcorrel = TSEDataUtils.calc_centered_moving_average(np.array(filtered_results_histcorrel), 10)
y_centered_moving_average_ed = TSEDataUtils.calc_centered_moving_average(np.array(filtered_results_ed), 10)   
y_centered_moving_average_histchisqr = TSEDataUtils.calc_centered_moving_average(np.array(filtered_results_histchisqr), 10)
y_centered_moving_average_corr = TSEDataUtils.calc_centered_moving_average(np.array(filtered_results_corr), 10) 
                                                                                                                                                                             
plt.plot(np.array(image_rows_ed), np.array(filtered_results_ed), "b.", label="DistanceEuclidean")
plt.plot(np.array(image_rows_corr), np.array(filtered_results_corr), "r.", label="DistanceCorr")
plt.plot(np.array(image_rows_histcorrel), np.array(filtered_results_histcorrel), "g.", label="HistCorrel")
plt.plot(np.array(image_rows_histchisqr), np.array(filtered_results_histchisqr), "y.", label="HistChiSqr")

plt.plot(image_rows_ed[len(image_rows_ed) - len(y_centered_moving_average_ed):], y_centered_moving_average_ed, "b-", label="DistanceEuclidean_CMA")
plt.plot(image_rows_corr[len(image_rows_corr) - len(y_centered_moving_average_corr):], y_centered_moving_average_corr, "r-", label="DistanceCorr_CMA")
plt.plot(image_rows_histcorrel[len(image_rows_histcorrel) - len(y_centered_moving_average_histcorrel):], y_centered_moving_average_histcorrel, "g-", label="HistCorrel_CMA")
plt.plot(image_rows_histchisqr[len(image_rows_histchisqr) - len(y_centered_moving_average_histchisqr):], y_centered_moving_average_histchisqr, "y-", label="HistChiSqr_CMA")                                                                                            
                                                                                                                                                      
plt.show()
```
